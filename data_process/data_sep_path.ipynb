```python
import pandas as pd
from tqdm import tqdm
import csv
import os

%pwd
```

```python
PATH = '../data/'
filenames = os.listdir(PATH+'/train')[:-4]

filenames
```

```python
for fn in filenames:
    with open(PATH+'train/'+fn, 'r') as f:
        head_part = [['order_id','ata','distance','simple_eta','driver_id','slice_id']]
        link_part = [['order_id','link_id','link_time','link_ratio','link_current_status','link_arrival_status','cross_flag']]
        for parts in tqdm(f.readlines()):
            line = parts.split(';;')
            head = line[0].split(' ')
            link = line[1].split(' ')
            order_id = head[0]
            head_part.append(head)
            k=0
            if(line[2]!='\n'):
                cross = line[2].split(' ')
                cross_no = len(cross)
                light = cross[k]
                id1 = light.split(':')[0].split('_')[0]
                id2 = light.split(':')[0].split('_')[1]    
            else:
                cross_no = 0   
                id1 = -1
                id2 = -1
            flag = 0
            path = []
            for i,segment in enumerate(link):
                tmp = [order_id]
                link_id = segment.split(':')[0]
                tmp.append(link_id)
                tmp.extend(segment.split(':')[1].split(','))
                tmp.extend([0])
                path.append(tmp)
                if(link_id==id1):
                    loc = i+1
                    flag = 1
                if(flag & (link_id==id2)):
                    path.insert(loc,[order_id,light.split(':')[0],light.split(':')[1],1,-1,-1,1])
                    flag = 0
                    if(k<cross_no-1):
                        k += 1
                        light = cross[k]
                        id1 = light.split(':')[0].split('_')[0]
                        id2 = light.split(':')[0].split('_')[1]
                    else:
                        id1 = -1
                        id2 = -1
            link_part.extend(path)
        with open(PATH+'train/train_head/'+fn[:8]+'.csv','w') as result_file:
            wr = csv.writer(result_file)
            wr.writerows(head_part)
        with open(PATH+'train/train_path/'+fn[:8]+'.csv','w') as result_file:
            wr = csv.writer(result_file)
            wr.writerows(link_part)
```
