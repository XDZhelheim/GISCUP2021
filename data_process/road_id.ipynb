## 2. 建图：下游路段是什么：不是邻接，也不是可达
---

```python
def convert(x):
    x=x.split(",")[0].split(":")
    x[0]=int(x[0])
    x[1]=float(x[1])

    return x

def check(line):
    roads=line[1].split(" ")
    roads=list(map(lambda x:convert(x), roads))
    unsorted=roads.copy()
    roads=sorted(roads, key=lambda x:x[0])

    cont=[]
    for i in range(len(roads)-1):
        if roads[i][0]+1==roads[i+1][0]:
            if i==0 or (i>=1 and not roads[i][0]-1==roads[i-1][0]):
                cont.append(roads[i])
            cont.append(roads[i+1])

    return unsorted, cont
```

```python
n=100000
lines=[]

with open("./data/train/20200801.txt", "r") as train_data:
    for _ in range(n):
        line=train_data.readline().split(";;")
        if len(line)>1:
            lines.append(line)
```

```python
for line in lines:
    _, res=check(line)
    if res:
        print(res)
```
