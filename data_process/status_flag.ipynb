```python
df=pd.read_csv("./link_status_attr.csv")
df.set_index("Unnamed: 0", inplace=True)

df
```

```python
def set_flag(row):
    cur_0=row["current_status_0"]
    cur_not_0=row[["current_status_1", "current_status_2", "current_status_3", "current_status_4"]].sum()
    if cur_0/(cur_0+cur_not_0)>=0.8:
        row["cur_flag"]=1
    else:
        row["cur_flag"]=0

    arr_0=row["arrival_status_0"]
    arr_not_0=row[["arrival_status_1", "arrival_status_2", "arrival_status_3", "arrival_status_4"]].sum()
    if arr_0/(arr_0+arr_not_0)>=0.8:
        row["arr_flag"]=1
    else:
        row["arr_flag"]=0

    # print(cur_0, cur_not_0, arr_0, arr_not_0)

    return row
```

```python
df=df.apply(set_flag, axis=1)

df
```
