```
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

os.getcwd()
```

```
file_number=10
count=0
df=None
filenames=os.listdir("../data/train/train_head")[1:] # skip 20200803
for f in filenames:
    if df is None:
        df=pd.read_csv(f"../data/train/train_head/{f}")
    else:
        df=df.append(pd.read_csv(f"../data/train/train_head/{f}"))

    count+=1
    if count>=file_number:
        break

df=df.dropna()
df
```

```
count_series=df["slice_id"].value_counts()
count_series=count_series.sort_index()
count_series
```

```
plt.figure(figsize=(30, 20))
count_series.plot(kind="bar", fontsize=5)
plt.show()
```
